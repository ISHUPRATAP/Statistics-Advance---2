QUESTION 1
Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps determine whether there is enough evidence in a sample to support a specific claim (called a hypothesis) about the population.

Key Concepts:

1-Null Hypothesis (H₀):
This is the default assumption or claim that there is no effect, no difference, or no relationship. For example, "The average height of students is 170 cm."

2-Alternative Hypothesis (H₁ or Ha):
This is what you want to prove — it suggests there is an effect, difference, or relationship. For example, "The average height of students is not 170 cm."

3-Test Statistic:
A value calculated from the sample data that is used to test the hypotheses (e.g., z-score, t-score).

4-P-value:
The probability of observing the sample data (or something more extreme) if the null hypothesis is true. A small p-value (typically ≤ 0.05) suggests that the null hypothesis may not be true.

5-Significance Level (α):
A threshold set by the researcher (commonly 0.05). If the p-value is less than α, the null hypothesis is rejected.

6-Conclusion:

If p-value ≤ α: Reject the null hypothesis (evidence supports the alternative).

If p-value > α: Fail to reject the null hypothesis (not enough evidence to support the alternative).

QUESTION 2
The null hypothesis (H₀) is a statement that there is no effect, no difference, or no relationship in the population — it represents the status quo or default assumption.

The alternative hypothesis (H₁ or Ha) is the opposite; it claims that there is an effect, a difference, or a relationship.

Key Differences:

H₀ assumes no change; H₁ suggests change.

H₀ is tested directly; H₁ is supported if H₀ is rejected.

The goal of hypothesis testing is to use sample data to determine whether to reject H₀ in favor of H₁.

QUESTION 3
The significance level (α) is the threshold used in hypothesis testing to determine whether to reject the null hypothesis (H₀). It represents the probability of making a Type I error, which is rejecting H₀ when it is actually true.

Common Values:

Typically set at 0.05, 0.01, or 0.10

Role in Hypothesis Testing:

After calculating the p-value, compare it to α:

If p-value ≤ α → Reject H₀ (evidence supports the alternative hypothesis)

If p-value > α → Fail to reject H₀ (not enough evidence to support the alternative)

QUESTION 4
Type I Error (False Positive):

Definition: Rejecting the null hypothesis (H₀) when it is actually true.

Probability: Equal to the significance level (α).

Example:
A medical test shows a patient has a disease (positive result), but they are actually healthy.
→ H₀: The patient does not have the disease (true)
→ Decision: Test says they do → Type I error

Type II Error (False Negative):

Definition: Failing to reject the null hypothesis when it is actually false.

Probability: Denoted by β.

Example:
A medical test shows a patient is healthy (negative result), but they actually have the disease.
→ H₀: The patient does not have the disease (false)
→ Decision: Test says they don’t → Type II error

Feature                     	Z-test	                        T-test
Population Std. Dev. (σ)	     Known	                       Unknown

Sample Size (n)              	Large (n > 30)	           Small (n ≤ 30)

Distribution Used	     Standard normal distribution (Z)	Student’s t-distribution

Purpose	 Test means or proportions in large samples	 Test means in small samples

When to Use Each:

Use a Z-test when:

The sample size is large (n > 30).

The population standard deviation (σ) is known.

The data is approximately normally distributed.

Use a T-test when:

The sample size is small (n ≤ 30).

The population standard deviation is unknown.

The data is approximately normal

```python
#QUESTION 7
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Parameters
mu_0 = 50               # Hypothesized population mean
sigma = 0.5             # Assumed population standard deviation
alpha = 0.05            # Significance level

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

```python
#QUESTION 8
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu = 100            # True mean
sigma = 15          # True standard deviation
n = 100             # Sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
confidence = 0.95
alpha = 1 - confidence
z_score = stats.norm.ppf(1 - alpha/2)

# Margin of error
margin_of_error = z_score * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#QUESTION 9
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Print basic stats
    print(f"Mean of original data = {mean:.2f}")
    print(f"Standard Deviation = {std_dev:.2f}")
    print(f"First 5 Z-scores: {z_scores[:5]}")

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.show()

    return z_scores

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=100)
z_scores = calculate_and_plot_z_scores(sample_data)
```
